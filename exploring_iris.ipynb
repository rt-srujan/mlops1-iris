```python
import pandas as pd
from sklearn.datasets import load_iris

data = load_iris(as_frame=True)
df = data.frame

print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
print("Summary stats:")
df.describe()
```

```python
print("Class counts:")
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

df.plot.scatter(
    x='sepal length (cm)',
    y='petal length (cm)',
    c='target',
    colormap='viridis'
)
plt.title('Iris dataset: see the clusters?')
plt.show()
```
